```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv(path+"train.csv")
test = pd.read_csv("../embeddings/use_embeddings_test.csv")
labels = pd.read_csv(path+"test_labels.csv")
```

```python
classes = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
y = train[classes].values
```

```python
X_t = pd.read_csv("../embeddings/use_embeddings.csv")
y_data = pd.DataFrame(y)
y_toxic = y_data[0]
y_severe_toxic = y_data[1]
y_obscene = y_data[2]
y_threat = y_data[3] 
y_insult = y_data[4]
y_identity_hate = y_data[5]
```

```python
def sigmoid(z):
    temp = 1+np.exp(-z)
    result = 1/temp
    return result
```

```python
def cost(theta, X, y):
    ''' cost function for you to minimize'''
    hx = sigmoid(np.dot(X, theta))
    first = y * np.log(hx)
    second = (1 - y) * np.log(1 - hx)
    return -np.sum(first +second) / len(X)
```

```python
def gradient(theta, X, y):
    theta_x = np.dot(theta, X.transpose())
    h_x = sigmoid(theta_x)
    grad = (1/len(y))*np.dot(h_x-y.transpose(), X)
    return grad
```

```python
# add 1:
X_t.insert(0, 'Ones', 1)
# initialize theta (0 initialization):
theta = np.array(np.zeros(513))
X_t.shape, theta.shape, y_toxic.shape
```

```python
#from sklearn.model_selection import train_test_split    #spilt train and validation set
#X_train, X_val, y_train, y_val = train_test_split(X_t,y,test_size=0.2,random_state=1)     #use the x,y of the original train
```

```python
#cost(theta, X_t, y_toxic)
cost(theta, X_t, y_toxic)
```

```python
import scipy.optimize as opt
```

```python
# train for y_toxic:
result1 = opt.minimize(fun=cost, x0=theta, args = (X_t, y_toxic), method='TNC', jac = gradient)
result1
```

```python
# train for y_severe_toxic:
result2 = opt.minimize(fun=cost, x0=theta, args = (X_t, y_severe_toxic), method='TNC', jac = gradient)
result2
```

```python
# train for y_obscene:
result3 = opt.minimize(fun=cost, x0=theta, args = (X_t, y_obscene), method='TNC', jac = gradient)
result3
```

```python
# train for y_threat:
result4 = opt.minimize(fun=cost, x0=theta, args = (X_t, y_threat), method='TNC', jac = gradient)
result4
```

```python
# train for y_insult:
result5 = opt.minimize(fun=cost, x0=theta, args = (X_t, y_insult), method='TNC', jac = gradient)
result5
```

```python
# train for y_identity_hate:
result6 = opt.minimize(fun=cost, x0=theta, args = (X_t, y_identity_hate), method='TNC', jac = gradient)
result6
```

```python
def predict(theta, X):
    #theta_x = np.dot(theta, X.transpose())
    #print(theta_x)
    probability = sigmoid(np.dot(theta, X.transpose()))
    return probability
```

```python
# print AUR-ROC score for the training set:
from sklearn.metrics import roc_auc_score, average_precision_score, accuracy_score

results = [result1, result2, result3, result4, result5, result6]
i = 0
scores_roc_aoc1 = []

print("AUC for the training set")
print("########################")
for label_name in classes:
    print('Class:', label_name)
 
    theta_min =  results[i].x
    i += 1
    preds_train = predict(theta_min, X_t)    # And we predict on the training set
    train_roc_auc = roc_auc_score(train[label_name], preds_train)
    print('Train ROC AUC Score:', train_roc_auc)
    scores_roc_aoc1.append(train_roc_auc)
    
print(np.mean(scores_roc_aoc1))# Comput the mean result of each column labels
```

```python
# print AUR-ROC score for the test set:
from sklearn.metrics import roc_auc_score, average_precision_score, accuracy_score

results = [result1, result2, result3, result4, result5, result6]
i = 0
scores_roc_aoc2 = []

labels = labels[classes]
sum_labels=np.sum(labels.values,axis=1)
idx=sum_labels>=0
y_test = labels[idx]
test.insert(0, 'Ones', 1)
X_test = test[idx]

print("AUC for the test set")
print("########################")
for label_name in classes:
    print('Class:', label_name)
 
    theta_min =  results[i].x
    i += 1
    preds_test = predict(theta_min, X_test)    # And we predict on the test set
    test_roc_auc = roc_auc_score(y_test[label_name], preds_test)
    print('Test ROC AUC Score:', test_roc_auc)
    scores_roc_aoc2.append(test_roc_auc)
    
print(np.mean(scores_roc_aoc2))# Comput the mean result of each column labels
```
